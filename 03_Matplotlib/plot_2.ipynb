```python
import pandas as pd
from collections import Counter
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("Data.csv")
lang=df["LanguagesWorkedWith"]
lang_count=Counter()
for x in lang:
    lang_count.update(x.split(";"))
languages=[]
counts=[]
for x in lang_count.most_common():
    languages.append(x[0])
    counts.append(x[1])
languages.reverse()
counts.reverse()

plt.barh(languages,counts)
plt.title("Popularity of programming languages")
plt.xlabel("Number of users")
plt.tight_layout()

plt.show()
```

```python
df=pd.read_csv("Data.csv")
lang_counter=df["LanguagesWorkedWith"].str.split(";").explode().value_counts()
lang_counter=lang_counter[::-1]
plt.barh(lang_counter.index,lang_counter.values)
plt.title("Popularity of programming languages")
plt.xlabel("Number of users")
plt.tight_layout()

plt.show()
```

```python
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("Data.csv")
df
```

```python
lang_count=df["LanguagesWorkedWith"].str.split(";").explode().value_counts() 
explode=[0,0,0,0.1,0]
plt.pie(lang_count.head(),labels=lang_count.head().index,autopct="%1.1f%%",shadow=True,explode=explode,wedgeprops={"edgecolor":"black"})
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv("data_4.csv")
df
```

```python
ages_x=df["Age"]
all_dev=df["All_Devs"]
py_dev=df["Python"]
js_dev=df["JavaScript"]


fig,(ax1,ax2)=plt.subplots(2,1,sharex=True)

ax1.plot(ages_x,all_dev,label="All Developers",color="black")

ax2.plot(ages_x,py_dev,label="Python Developers")
ax2.plot(ages_x,js_dev,label="JavaScript Developers")

ax1.legend()
ax1.set_title("Programming Languages Popularity")
ax1.set_ylabel("Popularity")
ax1.grid(True)

ax2.legend()
ax2.set_xlabel("Age Group")
ax2.set_ylabel("Popularity")
ax2.grid(True)

plt.tight_layout()
plt.show()
```

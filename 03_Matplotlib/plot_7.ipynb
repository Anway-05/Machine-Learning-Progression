```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("2019-05-31-data.csv")
df
```

```python
views_x=df["view_count"]
likes_y=df["likes"]
ratio=df["ratio"]

plt.style.use("seaborn-v0_8")

plt.scatter(views_x,likes_y,edgecolor="black",c=ratio,cmap="summer",s=50,alpha=0.7)

cbar=plt.colorbar()
cbar.set_label("Like_dislike ratio")

plt.title("Trending youtube videos")
plt.xlabel("View_Count")
plt.ylabel("Likes_Count")

plt.xscale("log")
plt.yscale("log")

plt.grid(True)
plt.tight_layout()
plt.savefig("plot.png", dpi=300,bbox_inches="tight")
plt.show()
```

```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df=pd.read_csv("btc.csv")
df
```

```python
df["Date"]=pd.to_datetime(df["Date"])
df.sort_values("Date",inplace=True)
date_x=df["Date"]
closing_value_y=df["Close"]

plt.plot(date_x,closing_value_y,marker='o',linewidth=3)
plt.gcf().autofmt_xdate()

plt.title("Bitcoin closing value trend")
plt.xlabel("Date")
plt.ylabel("Closing Value")

plt.grid(True)
plt.tight_layout()
plt.show()
```

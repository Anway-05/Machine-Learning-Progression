```python
a=np.ones((5,5))
a[1:-1,1:-1]=0
print(a)
```

```python
a=np.zeros((5,5))
x=[1,2,3,4]
for i in range(0,4):
    a[i,i+1]=x[i]
    a[i+1,i]=x[i]
print(a)
```

```python
a=np.random.rand(10,10)
blocks=[]
for i in range(0,8):
    for j in range(0,8):
        block=a[i:i+3,j:j+3]
        blocks.append(block)
print(len(blocks))
```

```python
a = np.array([[17,24,1,8,15],[23,5,7,14,16],[4,6,13,20,22],[10,12,19,21,3],[11,18,25,2,9]])
rows=np.sum(a,axis=0)
column=np.sum(a,axis=1)
print(max(rows.max(),column.max(),rows.min(),column.min(),np.sum(np.diag(a)),np.sum(np.diag(np.fliplr(a)))))
print(min(rows.max(),column.max(),rows.min(),column.min(),np.sum(np.diag(a)),np.sum(np.diag(np.fliplr(a)))))
```

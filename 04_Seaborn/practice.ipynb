```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("titanic")
df
```

```python
sns.countplot(data=df,x='class',hue='sex')
```

```python
sns.barplot(data=df,x='class',y='survived',hue='sex')
```

```python
sns.histplot(data=df,x='fare',kde=True,bins=5)
plt.yscale("log")
```

```python
sns.violinplot(data=df,y='fare',x="class")
plt.yscale("log")
```

```python
df1=sns.load_dataset("car_crashes")
df1
```

```python
sns.scatterplot(data=df1,x="alcohol",y="speeding")
```

```python
sns.jointplot(data=df1,x="alcohol",y="speeding",kind="reg")
```

```python
corr=df1.corr(numeric_only=True)
sns.heatmap(corr,annot=True,cmap="coolwarm")
```

```python
import pandas as pd
```

```python
orders = {
    "order_id": [1,2,3,4,5,6],
    "customer_id": [101,102,101,103,104,101],
    "product": ["Phone","Laptop","Phone","Tablet","Laptop","Tablet"],
    "quantity": [1,1,2,1,3,2]
}

customers = {
    "customer_id": [101,102,103,105],
    "name": ["Anway","Rahul","Lalo","Ghost"]
}

prices = {
    "product": ["Phone","Laptop","Tablet"],
    "price": [500,1200,300]
}
```

```python
df1=pd.DataFrame(orders)
df2=pd.DataFrame(customers)
df3=pd.DataFrame(prices)
```

```python
df1
```

```python
df2
```

```python
df3
```

```python
df4=pd.merge(df1,df2,on="customer_id",how="left")
df4
```

```python
df5=pd.merge(df3,df4,on="product")
df5
```

```python
df5["revenue"]=df5["price"]*df5["quantity"]
df5
```

```python
df5.pivot_table(index="customer_id",columns="product",values="revenue",aggfunc="sum")
```

```python
import pandas as pd
```

```python
df=pd.read_csv("Melbourne_housing_prices.csv")
df.head()
```

```python
cols_to_drop = ['Address','SellerG','Date','Postcode','CouncilArea','Lattitude','Longtitude','Method','Type','Regionname','Propertycount']
df = df.drop(columns=cols_to_drop)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df = df.dropna(subset=['Price'])
```

```python
df.isna().sum()
```

```python
num_cols = ['Distance','Bedroom2','Bathroom','Car','Landsize','BuildingArea','YearBuilt']
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum().sum()
```

```python
check_cols = ['Bedroom2','Bathroom','BuildingArea','YearBuilt']
for col in check_cols:
    print(col, (df[col] == 0).sum())
```

```python
df = df[(df['Bedroom2'] != 0) &(df['Bathroom'] != 0) &(df['BuildingArea'] != 0)]
df.shape
```

```python
df.describe()
```

```python
(df['YearBuilt'] < 1800).sum()
```

```python
df = df[df['YearBuilt'] >= 1800]
```

DATASET DESCRIPTION:

The Melbourne Housing dataset contains information about residential properties sold in Melbourne, Australia. The dataset contains property characteristics such as the number of rooms, property type, distance from the city center, land size, building area, year built, location details, and selling information.

After preprocessing and cleaning, the dataset contains approximately 27,000 observations and 10 columns, including the target variable Price.

```python
df['Suburb'].nunique()
```

DATASET OBJECTIVE:

The objective of this project is to predict house prices in Melbourne using the dataset. Multiple regression models, including Linear Regression, Ridge Regression, and Lasso Regression, will be used and compared to identify the model that provides the best predictions.

```python
x=df.drop('Price',axis=1)
y=df['Price']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder as ohe
from sklearn.preprocessing import StandardScaler as ss
from sklearn.compose import ColumnTransformer as ct
```

```python
c=ct([('Suburb',ohe(drop='first',handle_unknown='ignore'),['Suburb']),('numerical', ss(), x.columns.drop('Suburb'))])
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2,random_state=18)
```

```python
from sklearn.linear_model import LinearRegression as lr
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
model1=Pipeline([('preprocessor', c),('model', lr())])
model2=Pipeline([('preprocessor', c),('model', Lasso(alpha=1.0,max_iter=5000))])
model3=Pipeline([('preprocessor', c),('model', Ridge(alpha=1.0,max_iter=5000))])
```

```python
model1.fit(x_train,y_train)
model2.fit(x_train,y_train)
model3.fit(x_train,y_train)
```

MODEL COMPARISON:

Three regression models—Linear Regression, Ridge Regression, and Lasso Regression were used and compared on the Melbourne housing prices dataset. Linear Regression was used without regularization, while Ridge Regression used L2 regularization to reduce coefficient magnitudes and improve generalization. Lasso Regression used L1 regularization, which can shrink some coefficients to zero and perform feature selection.

```python
model1.score(x_train,y_train)
```

```python
model1.score(x_test,y_test)
```

```python
model2.score(x_train,y_train)
```

```python
model2.score(x_test,y_test)
```

```python
model3.score(x_train,y_train)
```

```python
model3.score(x_test,y_test)
```

FINDINGS & INTERPRETATION:

All three models—Linear Regression, Ridge Regression, and Lasso Regression—produced similar performance with an R² score of around 0.59. This indicates that regularization did not significantly impact model accuracy, suggesting limited overfitting in the dataset. The moderate score reflects that housing prices are only partially explained by the available features, and performance is likely constrained more by feature quality than model choice.

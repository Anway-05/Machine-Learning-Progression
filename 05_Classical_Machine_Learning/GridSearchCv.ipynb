```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split as tts
import pandas as pd
```

```python
dataset=load_digits(as_frame=True)
df=dataset.frame
df.head()
```

```python
x_train,x_test,y_train,y_test=tts(dataset.data,dataset.target,test_size=0.2,random_state=15)
```

```python
from sklearn.linear_model import LogisticRegression as lr
from sklearn.svm import SVC as svm
from sklearn.ensemble import RandomForestClassifier as rfc 
from sklearn.tree import DecisionTreeClassifier as dtc 
from sklearn.naive_bayes import GaussianNB as gnb
from sklearn.naive_bayes import MultinomialNB as mnb
from sklearn.model_selection import GridSearchCV as gsc
```

```python
model_params={
    'svm':{
        'model':svm(gamma='auto'),
        'params':{
            'C':[1,10,100],
            'kernel':["rbf","linear"]
        }
    },
    'random_forest': {
        'model': rfc(),
        'params' : {
            'n_estimators': [1,50,100]
        }
    },
    'logistic_regression' : {
        'model': lr(max_iter=1000),
        'params': {
            'C': [1,5,10]
        }
    },
    'decision_tree': {
        'model': dtc(),
        'params': {
            'criterion': ['gini', 'entropy'],
            'max_depth': [3, 5, 10, None]
        }
    },
    'gaussian_nb': {
        'model': gnb(),
        'params': {
            'var_smoothing': [1e-9, 1e-8, 1e-7]
        }
    },
    'multinomial_nb': {
        'model': mnb(),
        'params': {
            'alpha': [0.1, 0.5, 1.0, 2.0]
        }
    }
}
```

```python
scores=[]
for model,mp in model_params.items():
    c=gsc(mp["model"],mp["params"],cv=5)
    c.fit(x_train,y_train)
    scores.append({
        'model': model,
        'best_score': c.best_score_,
        'best_params': c.best_params_
    })
df1=pd.DataFrame(scores)
df1
```

```python
import pandas as pd
```

```python
df=pd.read_csv('height_weight.csv')
df
```

```python
df.describe()
```

```python
mean_height=df['height'].mean()
std_height=df['height'].std()
mean_weight=df['weight'].mean()
std_weight=df['weight'].std()
df['z_score_height']=((df['height']-mean_height)/std_height)
df['z_score_weight']=((df['weight']-mean_weight)/std_weight)
```

```python
q3_height=df['height'].quantile(0.25)
q5_height=df['height'].quantile(0.75)
q3_weight=df['weight'].quantile(0.25)
q5_weight=df['weight'].quantile(0.75)
```

```python
iqr_height=q5_height-q3_height
iqr_weight=q5_weight-q3_weight
```

```python
height_upper=q5_height+1.5*iqr_height
height_lower=q3_height-1.5*iqr_height
weight_upper=q5_weight+1.5*iqr_weight
weight_lower=q3_weight-1.5*iqr_weight
```

```python
df1=df[(df['z_score_height']>-3) & (df['z_score_height']<3) & (df['z_score_weight']>-3) & (df['z_score_weight']<3)]
df1.head()
```

```python
df2=df[(df['height']>height_lower) & (df['height']<height_upper) & (df['weight']>weight_lower) & (df['weight']<weight_upper)]
df2.head()
```

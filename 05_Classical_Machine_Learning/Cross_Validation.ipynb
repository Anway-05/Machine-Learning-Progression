```python
from sklearn.linear_model import LogisticRegression as lr
from sklearn.svm import SVC as svc
from sklearn.tree import DecisionTreeClassifier as dtc
from sklearn.ensemble import RandomForestClassifier as rfc
from sklearn.datasets import load_iris
from sklearn.model_selection import cross_val_score as cvs
```

```python
df=load_iris()
```

```python
scores_lr=cvs(lr(max_iter=1000),df.data,df.target,cv=10)
scores_lr.mean()
```

```python
scores_svc=cvs(svc(gamma="scale"),df.data,df.target,cv=10)
scores_svc.mean()
```

```python
scores_dtc=cvs(dtc(),df.data,df.target,cv=10)
scores_svc.mean()
```

```python
scores_svc=cvs(rfc(n_estimators=500),df.data,df.target,cv=10)
scores_svc.mean()
```

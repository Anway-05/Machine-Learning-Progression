```python
df1=df_dup_removed.copy()
df2=df_dup_removed.copy()
df2=df2.drop("Department",axis=1)
df3=df2.copy()
df3=df3.drop("Work_accident",axis=1)
```

```python
x1=df1.copy()
x1=x1.drop("left",axis=1)
y1=df1["left"]
x2=df2.copy()
x2=x2.drop("left",axis=1)
y2=df2["left"]
x3=df3.copy()
x3=x3.drop("left",axis=1)
y3=df3["left"]
```

```python
ct1=ColumnTransformer([("salary",OneHotEncoder(drop="first"),["salary"]),("Department",OneHotEncoder(drop="first"),["Department"])],remainder="passthrough")
x1=ct1.fit_transform(x1)
```

```python
ct2=ColumnTransformer([("salary",OneHotEncoder(drop="first"),["salary"])],remainder="passthrough")
x2=ct2.fit_transform(x2)
```

```python
ct3=ColumnTransformer([("salary",OneHotEncoder(drop="first"),["salary"])],remainder="passthrough")
x3=ct3.fit_transform(x3)
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y1,train_size=0.9,random_state=15)
x2_train,x2_test,y2_train,y2_test=train_test_split(x2,y2,train_size=0.9,random_state=15)
x3_train,x3_test,y3_train,y3_test=train_test_split(x3,y3,train_size=0.9,random_state=15)
```

```python
model1=lm.LogisticRegression(max_iter=10000)
model2=lm.LogisticRegression(max_iter=10000)
model3=lm.LogisticRegression(max_iter=10000)
```

```python
model1.fit(x1_train,y1_train)
```

```python
model2.fit(x2_train,y2_train)
```

```python
model3.fit(x3_train,y3_train)
```

```python
model1.score(x1_test,y1_test)
```

```python
model2.score(x2_test,y2_test)
```

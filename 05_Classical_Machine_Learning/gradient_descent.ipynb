```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
def gradient_descent(x,y):
    n=len(x)
    learning_rate=0.0002
    iterations=1000
    m_current=b_current=0
    for i in range(iterations):
        y_predicted=m_current*x+b_current
        mse=(1/n)*sum(pow(y-y_predicted,2))
        gradient_m=(2/n)*sum(x*(y-y_predicted))
        gradient_b=(2/n)*sum(y-y_predicted)
        if abs(gradient_m) < 0.001 and abs(gradient_b) < 0.001:
            print("Model converged")
            break
        m_current=m_current+learning_rate*gradient_m
        b_current=b_current+learning_rate*gradient_b
        print(f"For iteration {i+1} mse is {mse}, coeficient is {m_current} and intercept is {b_current}")
```

```python
df=pd.read_csv("test_scores.csv")
gradient_descent(df["math"],df["cs"])
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[["math"]],df[["cs"]])
```

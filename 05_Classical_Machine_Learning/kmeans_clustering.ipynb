```python
from sklearn.datasets import load_iris
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
```

```python
dataset=load_iris()
df=pd.DataFrame(dataset.data,columns=dataset.feature_names)
#df.drop(["petal length (cm)","petal width (cm)"],axis=1,inplace=True)
df
```

```python
ct=ColumnTransformer([("MinMax",MinMaxScaler(),["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)"])])
df=ct.fit_transform(df)
```

```python
model=KMeans(n_clusters=3)
y_predicted=model.fit_predict(df)
y_predicted
```

```python
dataset.target
```

```python
sse=[]
k_range=range(1,10)
for i in k_range:
    model1=KMeans(n_clusters=i)
    model1.fit(df)
    sse.append(model1.inertia_)
```

```python
plt.plot(k_range,sse)
```

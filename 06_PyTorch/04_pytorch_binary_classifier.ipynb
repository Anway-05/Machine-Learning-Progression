```python
x=torch.from_numpy(x).type(torch.float)
y=torch.from_numpy(y).type(torch.float)
```

```python
x,y
```

```python
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2,random_state=18)
```

```python
device="cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
class Binary_Classifier(nn.Module):
    def __init__(self):
        super().__init__()

        self.linear_layer1=nn.Linear(in_features=2,out_features=5)
        self.linear_layer2=nn.Linear(in_features=5,out_features=1)

    def forward(self,x):
        return self.linear_layer2(self.linear_layer1(x))
```

```python
torch.manual_seed(18)
model=Binary_Classifier()
model.state_dict()
```

```python
torch.manual_seed(18)
model_replica=nn.Sequential(
    nn.Linear(in_features=2,out_features=5),
    nn.Linear(in_features=5,out_features=1)
)
```

```python
model_replica.state_dict()
```

```python
loss_func=nn.BCEWithLogitsLoss()
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
epochs=10000
for _ in range(epochs):
    y_logit=model(x_train).squeeze()
    y_pred=torch.round(torch.sigmoid(y_logit))
    loss=loss_func(y_logit,y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
model.state_dict()
```

```python
model.eval()
with torch.inference_mode():
    y_test_logit=model(x_test).squeeze()
    y_test_pred=torch.round(torch.sigmoid(y_test_logit))
```

```python
accuracy_score(y_test,y_test_pred)
```

```python
class non_linear_binary_classifier(nn.Module):
    def __init__(self):
        super().__init__()

        self.layer1=nn.Linear(in_features=2,out_features=8)
        self.layer2=nn.Linear(in_features=8,out_features=8)
        self.layer3=nn.Linear(in_features=8,out_features=1)
        self.relu=nn.ReLU()

    def forward(self,x):
        return self.layer3(
            self.relu(
                self.layer2(
                    self.relu(
                        self.layer1(x)
                    )
                )
            )
        )
```

```python
torch.manual_seed(18)
non_linear_model=non_linear_binary_classifier()
non_linear_model
```

```python
loss_func=nn.BCEWithLogitsLoss()
optimizer=torch.optim.SGD(non_linear_model.parameters(),lr=0.05)
```

```python
epochs=10000
torch.manual_seed(18)
for _ in range(epochs):
    non_linear_model.train()
    y_logit=non_linear_model(x_train).squeeze()
    loss=loss_func(y_logit,y_train)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```

```python
non_linear_model.eval()
with torch.inference_mode():
    y_test_logit=non_linear_model(x_test).squeeze()
    y_test_pred=torch.round(torch.sigmoid(y_test_logit))
```

```python
accuracy_score(y_test,y_test_pred)
```

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.title("Train")
plot_decision_boundary(non_linear_model, x_train, y_train)
plt.subplot(1, 2, 2)
plt.title("Test")
plot_decision_boundary(non_linear_model, x_test, y_test)
```
